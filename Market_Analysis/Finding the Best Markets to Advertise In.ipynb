### <u>Checking for Sample Representativity</u>:
The courses offered by freeCodeCamp are mostly in web & mobile development but beside these main domains many others are covered as well which include data science, game development and many others. It turns out that only 31.7% of the participants have a clear idea about what programming niche they'd like to work in, while the vast majority of students have mixed interests. But given that we offer courses on various subjects, the fact that new coders have mixed interest might be actually good for us.

```python
#Boolean Values & frequency table
def _contains_(string):
    str_dev = na_values.str.contains(string)
    return str_dev

string_list = ['Full-Stack Web Developer', 'Front-End Web Developer', 
               'Back-End Web Developer', 'Web Developer', 'Mobile Developer', 
               'Data Scientist', 'Data Engineer', 'Game Developer']

freq_table_subject_dict = {}
for v in string_list:
    freq_table_subject_dict[v] = _contains_(v).value_counts(normalize = True) * 100

freq_table_subject_dict
```

```python
freq_table_subject = pd.DataFrame.from_dict(freq_table_subject_dict, orient = 'index')
freq_table_subject.reset_index(inplace = True)
freq_table_subject = freq_table_subject.rename(columns = {'index': 'Popular Subjects'})
freq_table_subject
```

```python
#fig, ax = plt.subplots(figsize=(10, 8))

freq_table_subject.plot.barh(figsize=(8,6))

plt.title('Most Popular Subjects Taught at Free Code Camp', 
          x = 0.51, y = 1.04, 
          fontweight = 'bold', 
          fontsize = 25) 

plt.xlabel('Percentage', fontsize = 13)

plt.yticks([0,1,2,3,4,5,6,7], 
           ['Full-Stack Web Developer', 'Front-End Web Developer', 'Back-End Web Developer', 'Web Developer', 
            'Mobile Developer', 'Data Scientist', 'Data Engineer', 'Game Developer'], 
           rotation = 0, fontsize = 12) 
plt.xlim([0,100])

# for location in ['left', 'right', 'top', 'bottom']:
#     plt.spines[location].set_visible(False)


plt.legend(bbox_to_anchor=[1.2, 1], loc='upper right')    
plt.tick_params(left = False, bottom = False)
sns.despine()
plt.show()
```

### <u>New Coders - Locations and Densities</u>:
Now that we found out that the sample has the right categories of people for our purpose, we can begin analyzing it. We can start with finding out where these new coders are located, and what are the densities (how many coders there are) for each location.

For our analysis, we'll work with the **`CountryLive`** variable because we're interested where people actually live at the moment when we run the ads. In other words, we're interested where people are located, not where they were born.

Because the data set provides information at a country level, we can think of each country as an individual market. This means we can frame our goal as finding the two best countries to advertise in.

To make sure we are working with a representative sample, we will drop the rows where participants didn't answer what role they are interested in. Where a participant didn't respond, we can't know for sure what their interests are, so it's better if we leave out this category of participants.

```python
participants = fcc['JobRoleInterest'].notnull().sum()
print(participants,'participants found with interest in a specific course of study.')

# Isolating participants that gave answer to the questions then those that didn't gave answers.
fcc_answered_participants = fcc[fcc['JobRoleInterest'].notnull()].copy()
```

```python
absolute_freq = fcc_answered_participants['CountryLive'].value_counts()
relative_freq = fcc_answered_participants['CountryLive'].value_counts(normalize = True) * 100
print(absolute_freq,'\n')
print(relative_freq)
```

**45.70 %** of our potential customers are located in the **USA**, and this definitely seems like the most **interesting market**. **India** has the **second customer density** but it's just **7.72 %** which means the market in India has **potential for further growth**. While markets in **UK (4.60 %)** & **Canada (3.80 %)** still has to evolve further.

we found useful information about the location of new coders, and what are the countries where new coders live, but we need to go more in depth with our analysis before taking a decision. We can continue by figuring out how much money new coders are actually willing to spend on learning. Advertising within markets where most people are only willing to learn for free is extremely unlikely to be profitable for us.
### <u>Spending Money for Learning</u>:
The **`MoneyForLearning`** column describes in American dollars the amount of money spent by participants from the moment they started coding until the moment they completed the course. The company sells subscriptions at a price of **$59 per month**, and for this reason we're interested in finding out how much money each student spends per month. 

It also seems like a good idea to narrow down our analysis to only four countries: **`the USA`**, **`India`**, **`the UK`**, and **`Canada`**. Two reasons for this decision are:

   - These are the countries having the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
   - Our courses are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our ads.


```python
fcc_answered_participants['MonthsProgramming'] = fcc_answered_participants['MonthsProgramming'].replace(0,1)
fcc_answered_participants['MoneySpent_perMonth'] = fcc_answered_participants['MoneyForLearning']/fcc['MonthsProgramming']
fcc_answered_participants['MoneySpent_perMonth'].isnull().sum()
```

```python
fcc_answered_participants['MoneySpent_perMonth'] = fcc_answered_participants[fcc_answered_participants['MoneySpent_perMonth'].notnull()]
fcc_answered_participants['CountryLive'].value_counts().head(4)
```

### <u>Defining Markets</u>:

```python
markets = fcc_answered_participants[fcc_answered_participants['CountryLive'].notnull()]
markets['CountryLive'].value_counts().head(4)
```

```python
markets['CountryLive'].value_counts().head(6).plot.barh(color = ['#d81159', '#8f2d56', '#218380', 
                                                                 '#fbb13c', '#a3f7b5', '#40c9a2'], 
                                                        figsize = (8,5))
sns.despine(left = True, bottom = True)
plt.gca().axes.get_xaxis().set_visible(False)
plt.tick_params(left = False, bottom = False)
plt.title('Potential Countries for Business', fontsize = 20, fontweight = 'bold', x = 0.12, y = 1.05)
plt.show()
```

```python
countries = ['United States of America', 'India', 'United Kingdom', 'Canada']
markets = markets[markets['CountryLive'].isin(countries)]
```

```python
markets_mean = markets.groupby('CountryLive')['MoneySpent_perMonth'].apply(np.mean)
markets_median = markets.groupby('CountryLive')['MoneySpent_perMonth'].apply(np.median)
print(markets_mean,'\n')
print(markets_median)
```

### <u>Dealing with Extreme Outliers</u>:

```python
sns.boxplot(y = 'MoneySpent_perMonth', x = 'CountryLive', data = markets)
plt.title('Money Spent per Month per Country\n(Distributions)', fontsize = 16)
plt.ylabel('Money Spent per Month (USD)')
plt.xlabel('Country')

# avoids tick labels overlap
plt.xticks(range(4), ['USA', 'UK', 'India', 'Canada']) 
plt.show()
```

We even can't see boxes. Let's try to cut the furthest outliers and take only values less than 59

```python
markets = markets[markets['MoneySpent_perMonth'] < 58]
markets_mean = markets.groupby('CountryLive')['MoneySpent_perMonth'].apply(np.mean)
markets_median = markets.groupby('CountryLive')['MoneySpent_perMonth'].apply(np.median)
```

```python
sns.boxplot(y = 'MoneySpent_perMonth', x = 'CountryLive', data = markets)
plt.title('Money Spent per Month per Country\n(Distributions)', fontsize = 16)
plt.ylabel('Money Spent per Month (USD)')
plt.xlabel('Country')

# Avoids tick labels overlap
plt.xticks(range(4), ['USA', 'UK', 'India', 'Canada'])
plt.show()
```

Still quite unclear. The number of outliers can tell us that participants could take into account money spent on primary univercity or bootcamps. Let's try to determine our potential customers as those, who spent between 10 and 40 USD per month and see how it affects on the size of our sample.

```python
condition = (markets['MoneySpent_perMonth'] > 10) & (markets['MoneySpent_perMonth'] < 40)
markets[condition]['MoneySpent_perMonth'].hist(grid = False, figsize = (8,5), color = '#ffb703', bins = 12)
plt.title('Money Spent per Month', fontsize = 25, fontweight = 'bold', x = 0.5)
plt.tick_params(axis='x', colors='#af0b1e')
plt.tick_params(axis='y', colors='#af0b1e')
sns.despine()
plt.show()
```

As we can see that the distribution is almost normal we may assume that we have saved the most important data when applying our condition. Let's see again how boxplot gives us representations this time.

```python
sns.boxplot(x = 'CountryLive',
            y = 'MoneySpent_perMonth', 
            data = markets[condition])

plt.title('Money Spent Per Month',fontsize = 16)
plt.ylabel('Money per month, USD')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'Canada', 'India'])
plt.show()
```

### <u>Results</u>:

```python
medians = markets[condition].groupby('CountryLive')['MoneySpent_perMonth'].apply(np.median)
means = round(markets[condition].groupby('CountryLive')['MoneySpent_perMonth'].apply(np.mean),4)
num_respondents = markets[condition]['CountryLive'].value_counts()
results = pd.DataFrame(data = {'Median': medians,
                                  'Mean': means,
                                  'Respondents': num_respondents})
results
```
